```python
import os
import mimetypes
import json

from typing import TypedDict, Optional, List, cast
```

```python
class IconResource(TypedDict):
  id: int
  name: str
  url: str
  format: str
  group: str
  tags: Optional[List[str]]
```

```python
resource_path = "../icons.json"

icons_folder_path = "../maki"
icons_family_name = "maki"
endpoint =  "https://raw.githubusercontent.com/giscaleing/metrix-icons/refs/heads/main"
```

```python
with open(resource_path, "r") as resource_file:
  resource = cast(List[IconResource], json.load(resource_file))
```

```python
icons_list = os.listdir(icons_folder_path)
icons_list.sort()
```

```python
for icon_item in icons_list:
  path = os.path.join(icons_folder_path, icon_item)

  identifier = len(resource) + 1
  name = os.path.splitext(os.path.basename(path))[0]
  mime_type, _ = mimetypes.guess_type(path)

  resource.append({
    'id': identifier,
    'group': icons_family_name,
    'name': name,
    'url': f"{endpoint}/{icons_family_name}/{icon_item}",
    'format': mime_type
  })
```

```python
with open(resource_path, "w", encoding="utf-8") as f:
  json.dump(resource, f, ensure_ascii=False, indent=2)
```
